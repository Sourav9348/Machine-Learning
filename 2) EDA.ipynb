Performing Exploratory Data Analysis(EDA) on Preprocessed crowdfunding Datasets.
Analysis includes:

1. Univariate Analysis
2. BiVariate Analysis
3. Multivariate Analysis

```python
# Importing necessary libraries

import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
warnings.filterwarnings('ignore')
```

```python
# Load the dataset

df = pd.read_csv("Cleaning data.csv")
df
```

```python
# Display the first few rows of the dataframe

df.head()
```

```python
#Dropping a column as there are two same column 

df.drop('Age_in_Days.1', axis=1, inplace=True)
df
```

```python
#Renaming some columns

df.rename(columns={'other.1': 'other_countries', 'other': 'other_categories'}, inplace=True)
df
```

```python
# Summary statistics of numerical columns
df.describe()
```

       Univariate analysis is the simplest form of analyzing the data. It is the analysis of only one feature from a set of different features. It normally gives us a description of a particular feature. It doesn’t deal with cause or relationship, it just takes the data, summarizes it, and represents it in the form of Histograms, Box and Whisker Plots, etc. We’ll be learning about this in the further sections.

```python
status_counts = df['status'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Company Status')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['funding_rounds'], bins=20, kde=True)
plt.title('Distribution of Funding Rounds')
plt.xlabel('Number of Funding Rounds')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y='funding_total_usd', data=df, color='skyblue')
plt.title('Distribution of Total Funding (USD)')
plt.ylabel('Total Funding (USD)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age_in_Days'], bins=30, kde=True, log_scale=True)
plt.title('Distribution of Company Age (in Days)')
plt.xlabel('Company Age (in Days)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='status', y='funding_total_usd', data=df)
plt.title('Distribution of Total Funding by Company Status')
plt.xlabel('Status')
plt.ylabel('Total Funding (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
country = df['USA'].value_counts()
plt.figure(figsize=(10, 6))
country.plot(kind='bar', color='salmon')
plt.title('Countries by Company Count')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for status in df['status'].unique():
    sns.kdeplot(df[df['status'] == status]['Age_in_Days'], label=status)
plt.title('KDE Plot of Company Age by Status')
plt.xlabel('Company Age (in Days)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
analytics_counts = df['analytics'].value_counts()
analytics_counts.plot(kind='bar', color='skyblue')
plt.title('Analytics Category Count')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

    Bivariate analysis explores the connection between two variables simultaneously. By examining how changes in one variable correspond to changes in another, we gain valuable insights into correlations, trends, and dependencies. Techniques like scatter plots and correlation coefficients help quantify these relationships, guiding decision-making in fields ranging from economics to healthcare. Bivariate analysis is a fundamental tool for uncovering meaningful connections and driving evidence-based decision-making.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age_in_Days', y='funding_total_usd', data=df, hue='status', palette='Set1')
plt.title('Scatter plot of Total Funding vs Company Age')
plt.xlabel('Company Age (Days)')
plt.ylabel('Total Funding (USD)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='status', y='funding_rounds', data=df, palette='pastel')
plt.title('Number of Funding Rounds by Company Status')
plt.xlabel('Status')
plt.ylabel('Number of Funding Rounds')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='status', y='funding_total_usd', data=df, palette='pastel')
plt.title('Distribution of Total Funding by Company Status')
plt.xlabel('Status')
plt.ylabel('Total Funding (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='status', y='Age_in_Days', data=df, palette='pastel')
plt.title('Age of Company by Status')
plt.xlabel('Status')
plt.ylabel('Age in Days')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='status', y='funding_total_usd', data=df, estimator=np.mean, ci=None, palette='Set2')
plt.title('Average Total Funding by Company Status')
plt.xlabel('Status')
plt.ylabel('Average Total Funding (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
selected_vars = ['funding_total_usd', 'Age_in_Days', 'funding_rounds', 'milestones']
sns.pairplot(df[selected_vars + ['status']], hue='status', palette='husl', diag_kind='kde')
plt.suptitle('Pairplot of Selected Variables', y=1.02)
plt.show()
```

```python
numeric_data = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

    Multivariate analysis delves into the intricate interplay among multiple variables simultaneously, providing a comprehensive understanding of complex data structures. Techniques like Principal Component Analysis (PCA), Factor Analysis, and Cluster Analysis uncover hidden patterns and dependencies, enabling informed decision-making across various disciplines. By exploring joint variations among variables, multivariate analysis unlocks deeper insights and enhances predictive accuracy, making it an invaluable tool for navigating intricate datasets and informing strategic actions.

```python
top_categories = df[['analytics', 'ecommerce', 'enterprise', 'games_video', 'software', 'web', 'health', 'biotech', 'cleantech', 'mobile']].apply(pd.Series.value_counts).T
top_categories['Total'] = top_categories.sum(axis=1)
top_categories = top_categories.sort_values(by='Total', ascending=False).head(10).drop('Total', axis=1)
plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar', stacked=True, cmap='tab10')
plt.title('Top 10 Product Categories by Company Status')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Status')
plt.show()
```

```python
top_countries = df[['USA', 'GBR', 'CAN', 'CHN', 'IND', 'FRA', 'ISR', 'NLD', 'SGP', 'RUS']].apply(pd.Series.value_counts).T
top_countries['Total'] = top_countries.sum(axis=1)
top_countries = top_countries.sort_values(by='Total', ascending=False).head(10).drop('Total', axis=1)
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', stacked=True, cmap='tab10')
plt.title('Top 10 Country Codes by Company Status')
plt.xlabel('Country Code')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Status')
plt.show()
```

```python
category_cols = ['analytics', 'biotech', 'cleantech', 'ecommerce', 'enterprise', 'games_video', 'hardware', 'health',
                 'medical', 'mobile', 'other_categories', 'social', 'software', 'web']
category_counts = df[category_cols].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
funding_over_time = df.groupby('founded_at')['funding_total_usd'].sum()
plt.figure(figsize=(12, 6))
funding_over_time.plot(kind='line', marker='o', color='green')
plt.title('Total Funding Over Time')
plt.xlabel('Year of Foundation')
plt.ylabel('Total Funding (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='funding_rounds', y='funding_total_usd', data=df, hue='status', palette='viridis')
plt.title('Scatter plot of Funding Rounds vs Total Funding')
plt.xlabel('Number of Funding Rounds')
plt.ylabel('Total Funding (USD)')
plt.show()
```

```python
selected_vars = ['funding_total_usd', 'Age_in_Days', 'milestones', 'relationships']
sns.pairplot(df[selected_vars + ['status', 'funding_rounds']], hue='status', palette='husl', diag_kind='kde')
plt.suptitle('Pairplot of Selected Variables with Funding Rounds', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='status', y='funding_total_usd', data=df, hue='analytics', split=True)
plt.title('Distribution of Total Funding by Status and Product Category')
plt.xlabel('Status')
plt.ylabel('Total Funding (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='status', y='Age_in_Days', data=df, hue='USA')
plt.title('Age of Company by Status and Country (USA)')
plt.xlabel('Status')
plt.ylabel('Age in Days')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
pivot_table = df.pivot_table(index='status', columns='USA', values='funding_total_usd', aggfunc='mean')
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mean Total Funding by Status and Country (USA)')
plt.xlabel('Country (USA)')
plt.ylabel('Status')
plt.show()
```

```python
numerical_cols = ['funding_total_usd', 'funding_rounds', 'milestones', 'relationships', 'Age_in_Days']
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.clustermap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Clustermap of Correlations Between Numerical Variables')
plt.show()
```

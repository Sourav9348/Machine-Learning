```python
x = df[train_features]
y = df['isClosed']
```

```python
# Importing necessary modules
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Splitting the data into training and testing sets, ensuring that the proportion of classes in the target variable 'y' is preserved (stratified sampling)
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.2)

# Standardizing the features using StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)  # Fitting the scaler to the training data and transforming it
x_test = scaler.fit_transform(x_test)    # Transforming the testing data using the same scaler fitted on the training data
```

```python
# Importing necessary modules for metrics computation and visualization
from sklearn.metrics import mean_squared_error, confusion_matrix, accuracy_score, precision_score, recall_score, classification_report, mean_absolute_error, f1_score
import itertools

# Definition of function to compute evaluation metrics
def metric(y_test, y_pred, x_train, y_train, model):
    # Computing test accuracy
    test_acc = accuracy_score(y_test, y_pred)
    
    # Computing train accuracy
    train_acc = model.score(x_train, y_train)
    
    # Computing precision score
    precision = precision_score(y_test, y_pred, average='macro')
    
    # Computing recall score
    recall = recall_score(y_test, y_pred, average='macro')
    
    # Computing mean squared error
    mse = mean_squared_error(y_test, y_pred)
    
    # Computing mean absolute error
    mae = mean_absolute_error(y_test, y_pred)
    
    # Computing F1_score
    F1_score = f1_score(y_test, y_pred, average='macro')
    
    # Printing computed metrics
    print('Test accuracy score:', test_acc, "%")
    print('Train accuracy score:', train_acc, "%")
    print('Precision score:', precision, "%")
    print('Recall score:', recall, "%")
    print('Mean Squared Error:', mse, "%")
    print('Mean Absolute Error:', mae, "%")
    print('F1-score:', F1_score, "%")
    
    # Printing classification report
    print("Classification report:")
    print(classification_report(y_test, y_pred))
    
    # Printing confusion matrix
    print("Confusion Matrix")
    conf_mat = confusion_matrix(y_test, y_pred)
    np.set_printoptions(precision=2)
    fig1 = plt.figure(figsize=(7, 6))
    plot_confusion_matrix(conf_mat, classes=y.value_counts().index.tolist(), title='Confusion matrix')
    
    # Returning computed metrics
    return test_acc, train_acc, precision, recall, mse, mae, F1_score

# Definition of function to plot confusion matrix
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Reds):
    # Plotting confusion matrix
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    plt.xticks(range(cm.shape[0])) 
    plt.yticks(range(cm.shape[1]))

    # Normalizing confusion matrix if required
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        cm = cm.round(2)
    else:
        cm = cm

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Initializing an empty list to store error rates
error_rate = []

# Looping through k values from 1 to 39
for i in range(1, 40):
    # Creating KNN classifier with the current value of k
    knn = KNeighborsClassifier(n_neighbors=i)
    
    # Fitting the classifier to the training data
    knn.fit(x_train, y_train)
    
    # Making predictions on the test data
    pred_i = knn.predict(x_test)
    
    # Converting y_test DataFrame to numpy array and flattening it
    y_test_ravel = y_test.to_numpy().ravel()
    
    # Calculating and storing the error rate for the current value of k
    error_rate.append(np.mean(pred_i != y_test_ravel))
```

```python
# Creating a plot figure with specified size
plt.figure(figsize=(10, 6))

# Plotting the error rate against different values of k
plt.plot(range(1, 40), error_rate, color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)

# Adding title and labels to the plot
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# Initializing an array to store mean accuracy for different values of k
mean_acc = np.zeros(20)

# Looping through different values of k from 1 to 20
for i in range(1, 21):
    # Creating KNN classifier with the current value of k and fitting it to the training data
    knn = KNeighborsClassifier(n_neighbors=i).fit(x_train, y_train)
    
    # Predicting labels for the test data
    yhat = knn.predict(x_test)
    
    # Computing and storing mean accuracy for the current value of k
    mean_acc[i - 1] = accuracy_score(y_test, yhat)

# Displaying the array of mean accuracy for each value of k
mean_acc
```

```python
# Generating x-axis ticks at integer intervals from 1 to 20
loc = np.arange(1, 21, step=1.0)

# Creating the plot figure with specified size
plt.figure(figsize=(10, 6))

# Plotting the mean accuracy against different values of k
plt.plot(range(1, 21), mean_acc)

# Setting x-axis ticks
plt.xticks(loc)

# Adding labels to the axes
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')

# Displaying the plot
plt.show()
```

```python
# Creating a KNN classifier with neighbor 1
knn = KNeighborsClassifier(n_neighbors=1)

# Fitting the classifier to the training data
knn.fit(x_train, y_train)

# Making predictions on the test data
knn_predictions = knn.predict(x_test)
```

```python
# Computing various evaluation metrics using the 'metric' function
test_acc_knn, train_acc_knn, precision_knn, recall_knn, mse_knn, mae_knn, f1_score_knn = metric(y_test, knn_predictions, x_train, y_train, knn)
```

```python
model_result = pd.DataFrame({'test accuracy':test_acc_knn , 'train accuarcy':train_acc_knn,
                            'precision score':precision_knn , 
                            'recall score':recall_knn,
                            'mse':mse_knn,
                            'mae':mae_knn,
                            'F_1 score':f1_score_knn}  , index=['KNN_1'])
model_result
```

```python
# Importing GridSearchCV from scikit-learn
from sklearn.model_selection import GridSearchCV

# Defining the grid of hyperparameters for grid search
leaf_size = list(range(1, 50))
n_neighbors = [1, 2]
p = [1, 2]
hyperparameters = dict(leaf_size=leaf_size, n_neighbors=n_neighbors, p=p)

# Initializing GridSearchCV with the KNN classifier, grid of hyperparameters, and cross-validation strategy
clf = GridSearchCV(knn, hyperparameters, cv=3, n_jobs=-1)

# Flattening the y_train DataFrame if needed
y_train_ravel = y_train.to_numpy().ravel()


# Fitting the grid search to the training data
best_model = clf.fit(x_train, y_train_ravel)

# Obtaining the best hyperparameters
best_params = best_model.best_params_
print("Best hyperparameters:", best_params)
```

```python
# Printing the best hyperparameters found by the grid search
print('Best leaf_size:', best_model.best_estimator_.get_params()['leaf_size'])
print('Best p:', best_model.best_estimator_.get_params()['p'])
print('Best n_neighbors:', best_model.best_estimator_.get_params()['n_neighbors'])
```
